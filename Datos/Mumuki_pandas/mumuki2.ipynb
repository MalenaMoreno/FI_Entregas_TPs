```python
import pandas as pd 
```

```python
arboles = pd.read_csv("https://github.com/MumukiProject/datasets/raw/master/arbolado-publico-lineal.csv")
arboles.head(5)
```

```python
arboles["height"] >= 7
```

```python
arboles[arboles["height"] >= 7] #Todos los árboles que miden 7 o más metros de alto
```

```python
arboles[arboles["neighbourhood"] == "PALERMO"] #Todos los árboles del barrio de Palermo.
```

```python
arboles[arboles["comm_name"].str.startswith("Sauce")] #Todos los sauces
```

```python
#Podemos filtrar las dos condiciones al mismo tiempo utilizando la conjunción lógica &

#Sauces de palermo 

arboles[(arboles["neighbourhood"] == "PALERMO") & arboles["comm_name"].str.startswith("Sauce")]
```

```python
de_4_barrios = arboles[
     (arboles["neighbourhood"] == "FLORESTA") 
     | (arboles["neighbourhood"] == "RECOLETA") 
     | (arboles["neighbourhood"] == "BELGRANO")
     | (arboles["neighbourhood"] == "NUÑEZ")]

len(de_4_barrios)
```

```python
de_4_barrios2 = arboles[arboles["neighbourhood"].isin(["FLORESTA", "RECOLETA", "BELGRANO", "NUÑEZ"])]

len(de_4_barrios2)
```

```python
arboles[~arboles["comm_name"].str.startswith("Fresno")]
```

```python
sin_infor = arboles[arboles["street"].isna()]

len(sin_infor)
```

```python
valor_no_nulo = arboles[arboles["num1"].notna()]

len(valor_no_nulo)
```
